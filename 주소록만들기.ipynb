```python
import json

# 주소록 데이터 (초기값)
addressbook = {
    "김영호": "01047278728",
    "임세진": "01093413396"
}

while True:
    name = input("이름을 입력하세요 (끝 입력 시 종료): ")

    if name == "끝":
        break

    phoneNum = input("전화번호를 입력하세요: ")
    addressbook[name] = phoneNum

# 파일로 저장
with open("addressbook.json", "w", encoding="utf-8") as f:
    json.dump(addressbook, f, ensure_ascii=False, indent=4)

print("주소록이 addressbook.json 파일로 저장되었습니다!")

#이거는 기존에 쓴 주소록이 새롭게 작성하면 날라감
#기존에 있는 파일 불러와서 추가할수 있게 or 탐색기 더해보기
```
