```python
import json
import os # os는 운영체제 관련기능 사용하기 위해 불러온 모듈 ,, 파일 존재 경로 등 확인

# 주소록 데이터 (초기값)
addressbook = {
    "김영호": "01047278728",
    "임세진": "01093413396"
}

# 기존 파일 불러오기
if os.path.exists("addressbook.json"):
    with open("addressbook.json", "r", encoding="utf-8") as f:
        addressbook = json.load(f)
    print("기존 주소록을 불러왔습니다!")
else:
    addressbook = {
        "김영호": "01047278728",
        "임세진": "01093413396"
    }
    print("새 주소록을 생성했습니다!")

while True:
    print("\n[1] 추가  [2] 전체출력  [3] 삭제  [4] 종료")
    choice = input("선택: ")

    if choice == "1":
        name = input("이름을 입력하세요: ")
        phoneNum = input("전화번호를 입력하세요: ")
        addressbook[name] = phoneNum
        print(f"{name} 추가완료!")

    elif choice == "2":
        if not addressbook:
            print("주소록이 비어있습니다.")
        else:
            print("\n===== 주소록 =====")
            for name, phone in addressbook.items():
                print(f"{name} : {phone}")

    elif choice == "3":
        name = input("삭제할 이름을 입력하세요: ")
        if name in addressbook:
            del addressbook[name]
            print(f"{name} 삭제완료!")
        else:
            print("해당 이름이 없습니다.")

    elif choice == "4":
        break

# 파일로 저장
with open("addressbook.json", "w", encoding="utf-8") as f:
    json.dump(addressbook, f, ensure_ascii=False, indent=4)

print("주소록이 addressbook.json 파일로 저장되었습니다!")

```
