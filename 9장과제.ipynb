```python
# 1.
# a) 객체 지향 프로그래밍: 프로그램을 객체들의 집합과 상호작용으로 모델링하여 개발하는 기법
# b) 절차적 프로그래밍: 함수나 모듈을 문제해결 순서에 맞게 호출하여 수행하는 방식
# c) 그래픽 사용자 인터페이스: 사용자가 그래픽 요소(버튼, 텍스트필드 등)를 통해 컴퓨터와 상호작용하는 환경

# 2. 절차적 프로그래밍은 데이터와 함수가 분리되어 있고 순서대로 호출하는 방식 객체 지향 프로그래밍은 속성과 메소드를 하나의 객체로 묶어 객체 간 상호작용으로 프로그램을 구성한다.
```

```python
class Dog:
    def __init__(self, name):
        self.name = name

    def bark(self):
        print("멍멍~~")

my_dog = Dog('Jindo')
my_dog.bark()
```

```python
class Dog:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return 'Dog(name = ' + self.name + ')'

my_dog = Dog('Jindo')
print('my_dog의 정보 :', my_dog)
```

```python
class Vector2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __mul__(self, other):
        return Vector2D(self.x * other.x, self.y * other.y)

    def __truediv__(self, other):
        return Vector2D(self.x / other.x, self.y / other.y)

    def __neg__(self):
        return Vector2D(-self.x, -self.y)

    def __str__(self):
        return "({}, {})".format(self.x, self.y)

v1 = Vector2D(30, 40)
v2 = Vector2D(10, 20)
print('v1 * v2 =', v1 * v2)
print('v1 / v2 =', v1 / v2)

v1 = Vector2D(10, 20)
print('-v1 =', -v1)
```

```python
import math

class Vector2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def magnitude(self):
        return math.sqrt(self.x ** 2 + self.y ** 2)

    def __gt__(self, other):
        return self.magnitude() > other.magnitude()

    def __ge__(self, other):
        return self.magnitude() >= other.magnitude()

    def __lt__(self, other):
        return self.magnitude() < other.magnitude()

    def __le__(self, other):
        return self.magnitude() <= other.magnitude()

v1 = Vector2D(30, 40)
v2 = Vector2D(10, 20)
print('v1 > v2 =', v1 > v2)
print('v1 >= v2 =', v1 >= v2)
print('v1 < v2 =', v1 < v2)
print('v1 <= v2 =', v1 <= v2)
```
