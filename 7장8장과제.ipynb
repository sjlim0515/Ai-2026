```python
import datetime as dt

today = dt.date.today()
print('오늘의 날짜 : {}년,{}월{}일'.format(today.year,today.month,today.day))
start_time = dt.datetime.now()
hour = start_time.hour
if hour >12:
    print('오후',hour-12,'시')
elif hour==12:
    print('오후',hour,'시')
else:
    print ('오전',hour,'시')
print('현재시간: {}시 {}분 {}초'.format(hour,start_time.minute,start_time.second))
```

```python
import datetime as dt

today = dt.datetime.today()

print('오늘은 {}년{}월{}일 입니다'.format(today.year,today.month,today.day))

Xmas = dt.datetime(2026,12,25)
timegap = Xmas - dt.datetime.now()

print ('크리스마스때까지 {}일{}시간 남았습니다'.format(timegap.days,timegap.seconds//3600))

newyear = dt.datetime(2027,1,1)
timegap2 = newyear - dt.datetime.now()
print ('새해까지 {}일{}시간 남았습니다'.format(timegap2.days, timegap.seconds//3600))

birthday = dt.datetime(2026,5,15)
timegap3 = birthday - dt.datetime.now()
print('생일까지 {}일{}시간 남았습니다'.format(timegap3.days, timegap.seconds//3600))
```

```python
import datetime as dt

year = int(input("연도 입력: "))
month = int(input("월 입력: "))
day = int(input("일 입력: "))

start = dt.date(year, month, day)
result = start + dt.timedelta(days=100)

print("{}년 {}월 {}일 로부터 100일 후:".format(year,month,day))
print("{}년 {}월 {}일".format(result.year,result.month,result.day))
```

```python
import math as m

for i in range(2,11):
    print('4**',i,'=',m.pow(4,i))

#m.pow(x,y) x의 y 제곱
#m.fabs(-99) -99의 실수 절댓값
#m.ceil(2.1) = 2.1의 올림값
#m.ceil(-2.1) = -2.1의 올림값
#m.floor(2.1) = 내림값
#n.log(100,10) = log100 

for i in range(0,181,10):
    print(i,'degree=',round(m.radians(i),3),'radian')

for i in range (0,181,10):
    print('sin(',i,')=', round(m.sin(m.radians(i)),2))
```
