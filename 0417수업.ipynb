```python
class Student:
    def __init__(self, name, score):
        self.name = name
        self.score = score
    def get_grade(self):
        if self.score >= 90:
            return "A"
        elif self.score >= 80:
            return "B"

민수 = Student("민수",85)
지영 = Student("지영",92)

print (민수.get_grade())
print (지영.get_grade())
```

```python
class Phone:
    def __init__(self, brand, battery):
        self.brand = brand
        self.battery = battery
    
    def use(self, minutes):
        self.battery -= 0.5 * minutes
        print('브랜드[{}], 배터리 {}분 사용 > 잔량: {}%'.format(self.brand, minutes, self.battery))
    
    def charge(self, minutes):
        self.battery += 0.5 * minutes
        print('브랜드[{}], 배터리 {}분 충전 > 잔량: {}%'.format(self.brand, minutes, self.battery))

my_Phone = Phone("Galaxy", 80)
#my_Phone.brand = 'Galaxy' , my_Phone.battery = 80
my_Phone.use(30)
my_Phone.charge(20)
```
