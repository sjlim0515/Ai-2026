```python
s= 0
for i in range(1,11): # 1 = 1 , 11= 10
    s = s+i #반복문에 의해 0+1 > s=1 1+2 > s=2 .....> s=9 9+10 = 55

print ('1에서 10까지의 합:', s)
```

```python
s= 0
for i in range(1,11): # 1 = 1 , 11= 10
    s = s+i
    print('i={}, s={}'.format(i,s)) #. 들어가면 뒤가 매소드?? 임..  format 통해 {},{}가 i, s 라는걸 넣음 
print('1에서 10까지의 합:',s)
```

```python
n= int(input('합계를 구할 수를 입력해라:'))
s = 0

for i in range(0,n):
    s = s+(i+1) #i+1인 이유 ~> range n 하면 n-1 까지 생성하기 때문

print ('1부터 {}까지의 합은 {}'.format(n,s)) #{},{} 는 n,s라는것
```
